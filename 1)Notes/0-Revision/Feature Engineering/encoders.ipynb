```python
import seaborn as sns
```

```python

tips = sns.load_dataset('tips')
```

```python
tips
```

```python
import pandas as pd
```

```python
sex=pd.DataFrame(tips['sex'])
```

```python
sex
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
sex_table=encoder.fit_transform(sex).toarray()
```

```python
sex_table
```

```python
sex_df=pd.DataFrame(sex_table,columns=encoder.get_feature_names_out())
```

```python
sex_df
```

```python
pd.concat([sex,sex_df],axis=1)
```
