Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
df = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
df.head()
```

```python
df.info()
```

Data Cleaning

```python
#missing values

df[df.isnull().any(axis=1)]
```

```python
#This dataset is divided into two region from the index 122th

df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Removing the null values 

df=df.dropna().reset_index(drop=True)

df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# WE have to also drop this row because of repition of head row

df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# there is spaces in coloumn names so fix it

df.columns=df.columns.str.strip()
df.columns
```

```python
# change the required columns as integers
df[['day', 'month', 'year', 'Temperature', 'RH','Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH','Ws']].astype(int)
```

```python
df.info()
```

```python
# changning the other features into float

objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
# lets save the clean data into new csv file

df.to_csv("Algerian_forest_fires_dataset_Cleaned.csv",index=False)
```
