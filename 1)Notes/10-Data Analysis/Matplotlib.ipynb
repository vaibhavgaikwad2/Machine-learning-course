```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Create a line plot
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Creating customized line plot

plt.plot(x,y,color='green',linestyle='--')
```
