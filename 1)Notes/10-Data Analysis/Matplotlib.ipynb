```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Create a line plot
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Creating customized line plot

plt.plot(x,y,color='green',linestyle='--')
```

```python
# using different linestyle

plt.plot(x,y,linestyle='-.')
```

```python
plt.plot(x,y,linestyle=':')
```

```python
# Marker paarmeter used to locate co-ordinates
plt.plot(x,y,color='green',linestyle='-',marker='o')
```

```python
# Linewidth --> increases thickness of line 

plt.plot(x,y,linewidth='7')
```

```python
plt.plot(x,y)
plt.grid(True)
```

```python
# creating multiple plots

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]
plt.Figure(figsize=(9,5))

plt.subplot(2,2,1) # --> 1 row nd 1 coloumn and last parameter is for how much graphs you ploting
plt.plot(x,y1,color='green')
plt.title('plot 1')

plt.subplot(2,2,2) # --> 1 row nd 1 coloumn
plt.plot(y1,x,color='blue')
plt.title('plot 2')

plt.subplot(2,2,3) # --> 1 row nd 1 coloumn
plt.plot(x,y2,color='black')
plt.title('plot 3')


plt.subplot(2,2,4)
plt.plot(y2,x,color='red')
plt.title('Plot 4')
```

# Pie Charts

```python
labels = ['A','B','C','D']
sizes = [30,20,40,10]
colors=['gold','yellow','green','red']
explode=(0.2,0,0,0) # --> used to moveout the first slice

plt.pie(sizes,labels=labels)
```

```python
plt.pie(sizes,labels=labels,colors=colors)
```

```python
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%")
```

```python
plt.pie(sizes,labels=labels,colors=colors,explode=explode,shadow=True)
```

```python
# Sales Data Visualization

import pandas as pd
sales_data=pd.read_csv('sales_data.csv')
print(sales_data.head())
```

```python
## Plot total sales by products
sales_data.info()  #--> this gives info about data
```

```python
total_sales_by_product=sales_data.groupby('Product Category')['Total Revenue'].sum()
```

```python
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind='bar',color='red')
```

```python
## plot sales trend over time

sales_trend=sales_data.groupby('Date')['Units Sold'].sum().reset_index()
```

```python
sales_trend
```

```python
plt.plot(sales_trend['Date'],sales_trend['Units Sold'])
```

```python
sales_trend=sales_data.groupby('Date')['Total Revenue'].sum().reset_index() 

# --> by doing reset index there will be only two cloumns

plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
