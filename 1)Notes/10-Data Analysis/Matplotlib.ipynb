```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Create a line plot
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Creating customized line plot

plt.plot(x,y,color='green',linestyle='--')
```

```python
# using different linestyle

plt.plot(x,y,linestyle='-.')
```

```python
plt.plot(x,y,linestyle=':')
```

```python
# Marker paarmeter used to locate co-ordinates
plt.plot(x,y,color='green',linestyle='-',marker='o')
```

```python
# Linewidth --> increases thickness of line 

plt.plot(x,y,linewidth='7')
```

```python
plt.plot(x,y)
plt.grid(True)
```

```python
# creating multiple plots

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]
plt.Figure(figsize=(9,5))

plt.subplot(2,2,1) # --> 1 row nd 1 coloumn and last parameter is for how much graphs you ploting
plt.plot(x,y1,color='green')
plt.title('plot 1')

plt.subplot(2,2,2) # --> 1 row nd 1 coloumn
plt.plot(y1,x,color='green')
plt.title('plot 2')

plt.subplot(2,2,3) # --> 1 row nd 1 coloumn
plt.plot(x,y2,color='green')
plt.title('plot 3')

```
