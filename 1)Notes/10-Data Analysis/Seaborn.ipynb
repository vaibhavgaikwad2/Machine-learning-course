```python
import seaborn as sns
```

```python
# basic plotting with seaborn

tips=sns.load_dataset('tips')  # --> there is a lot of inbuilt datasets are available
tips
```

# Scatter Plot

```python
# Creating a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Bill v/s Tip")
```

# Line Plot

```python
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Size v/s total bill line plot')
```

# Categorical plots


```python
# Bar plot

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of total bill v/s day')
```

```python
# Box plot

sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box plot of total bill v/s day')
```

```python
# Violin plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# histogrmas

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
# Kde plot

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
correlation = tips[['total_bill','tip','size']].corr()
correlation
```

```python
sns.heatmap(correlation,annot=True,cmap='coolwarm')
```
